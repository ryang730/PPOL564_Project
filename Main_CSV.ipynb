```python
import requests
import urllib
import pandas as pd
import country_converter as coco
from bs4 import BeautifulSoup #for parsing the website
import time #for system to sleep 
import random
from dfply import *
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Url that we'll scrape. 
## notice there are several pages of the table so we will need to make a list to scrape table from this website
## 35 pages in total
ns_url_list = list()
ns_url_head = "https://www.vgchartz.com/games/games.php?page="
ns_url_tail = "&console=NS&order=Sales&ownership=Both&direction=DESC&showpublisher=1&showreleasedate=1&showlastupdate=1&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1"
page_list = list(range(1,36))
for i in page_list:
    ns_url = ns_url_head + str(i) + ns_url_tail
    ns_url_list.append(ns_url)
#ns_url_list
## after gathered all ns_url_list, we get the full table

```

```python
ns_url_list
```

```python
df_list = list()
for link in ns_url_list:
    time.sleep(random.uniform(2,10))
    page = requests.get(link)
   
    table = pd.read_html(page.text)
    df = table[6]
    df_list.append(df)
   
```

```python
df_list[1]
```

```python
new_df_list = list()
for dat in df_list:
    df = dat.set_axis(['Pos', 'Game', 'Game.1', 'Console', 'Publisher','VGChartz Score','Critic Score','User Score','Total Shipped','Release Date','Last Update','random'], axis=1, inplace=False)
    new_df_list.append(df)
   
```

```python
main_df = pd.concat(new_df_list,ignore_index=True)
main_df = main_df.filter(items = ['Game.1','Publisher','VGChartz Score','Critic Score','User Score','Total Shipped','Release Date'])
# change the release date to year
year_value = main_df['Release Date'].str.split(" ", n = 2, expand = True)
main_df['Release_year'] = year_value[2]

main_df.head()
```
